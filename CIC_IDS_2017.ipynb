```python
import os
import pandas as pd
import numpy as np
# loads the env
from dotenv import load_dotenv
import kagglehub

# 1. Load the  env file
load_dotenv()

# 2. Assign credentials (FIXED: KAGGLE_KEY now maps to the correct variable)
os.environ['KAGGLE_USERNAME'] = os.getenv('KAGGLE_USERNAME', '')
os.environ['KAGGLE_KEY'] = os.getenv('KAGGLE_KEY', '')

import kaggle 

# 4. Attempting Authentication
try:
    kaggle.api.authenticate()
    print("Successfully obtained env/creds :)")
except Exception as e:
    print(f"Authentication unfortunately failed :( : {e}")

path = kagglehub.dataset_download("chethuhn/network-intrusion-dataset")
print(" Path to dataset files:", path)
```
